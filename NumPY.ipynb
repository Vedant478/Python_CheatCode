```python
#accept password and rate it
#+1 if a-z
#+1 if A-Z
#+1 if 0-9
#+1 if special symbol used

#4 is strong  3 is good 2 is weak rejected
```

```python
#accept passwrod and rate it
#+1 if a-z
#+1 if A-Z
#+1 if 0-9
#+1 if special symbol used
#4 is strong 3 is good 2,1 weak rejected
import re
password = input("Enter a password: ")
if len(password)>=8:
    score = 0
    if re.search(r'[a-z]', password):
        score += 1
    if re.search(r'[A-Z]', password):
        score += 1
    if re.search(r'[0-9]', password):
        score += 1
    if re.search(r'\W', password):
        score += 1

    if score == 4:
        print("Strong password")
    elif score == 3:
        print("Good password")
    elif score == 2 or score == 1:
        print("Weak password")
    else:
        print("Rejected password")
else:
    print("Rejected password due to length")
```

```python
s1 = 0
for item in ls_dob:
    if item.isdigit():
        s1 += int(item)

print("Initial sum of digits:", s1)


temp = s1
while temp > 9:
    new = 0
    while temp > 0:
        digit = temp % 10
        new += digit
        temp //= 10
    temp = new

print("Final single digit:", temp)


print("Personality trait is:", personality_traits[temp])
```
