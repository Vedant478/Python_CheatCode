```python
# accept n numbers till 1st blank
#print sum and average and min numbers and max number entered

sum =0
count=0
avg=0
min = 9999 #start from largest
max = -9999 #start from smallest
while True:
  data = input("Enter number")
  if data=='':
    print("Ok bye", sum)
    print(count)
    avg = sum / count
    print(avg)

    break
  else:
    no=int(data)
    count+=1
    if no<min:
      min=no
    if no>max:
      max=no
    sum+=int(data)

print(f"Mininmum: {min} and Maximum: {max}")
```

```python
#Accept n numbers till blank
#count numbers perfectly div by 3,5,7 individually
#also all numbers perfectly divisible by 3 and 5
#also all numbers perfectly divisible by 5 and 7

sum =0
count3=0
count5=0
count7=0
counttf=0
countfs=0
while True:
  data = input("Enter number")
  if (data==""):
    print("Count by 3",count3)
    print("Count by 5",count5)
    print("Count by 7",count7)
    print("Count by 3/5",counttf)
    print("Count by 5/7",countfs)
    break
  else:
    n = int(data)
    if n%3==0:
      count3+=1
    if n%5==0:
      count5+=1
    if n%7==0:
      count7+=1
    if (n%3==0 and n%5==0):
      counttf+=1
    if (n%5==0 and n%7==0):
      countfs+=1
```

```python
# Problem Definition: Guess the Price of a Product
# Objective: Write a program where the user enters the name and price of a product, and the player has to guess the price. The program should provide feedback on whether the guess is too high, too low, or correct. The guessing continues until the correct price is guessed.

# Requirements:
# The program should prompt the user to input the name of the product and its price.
# The player should be able to guess the price of the product.
# The program should inform the player if their guess is too high, too low, or correct.
# The guessing should continue indefinitely until the correct price is guessed.
# Input:
# Product Name: A string representing the name of the product.
# Product Price: A float or integer representing the price of the product.
# Player's Guess: A float or integer representing the guessed price.
# Output:

# Feedback on whether the guess is too high, too low, or correct.
# Example Input and Output:
# Input:
# Enter the product name: Coffee
# Enter the product price: 5.99
# Guess the price of Coffee: 4.50
# Too low! Try again.
# Guess the price of Coffee: 6.00
# Too high! Try again.
# Guess the price of Coffee: 5.99
# Correct! You've guessed the price of Coffee.
```

```python
P_name=input("Enter a product name")
c_price = float(input("Enter the actual price"))
count =5
flag = False
while True and count>0:
  P_price = float(input(f"Guess the price: {P_name} "))
  count-=1
  if(P_price==c_price):
    print("You are the goat: ")
    flag=True
    break
  if(P_price<c_price):
    print(f"Too low! {count} try again: ")
  if(P_price>c_price):
    print(f"Too high! {count} Try again: ")
if flag==False:
  print("You lost it")
else :
  print("You won it")
```
